```python
import pandas as pd

df = pd.read_csv("toi179_authors.csv", encoding='unicode_escape')
df.head()
```

```python
len(set(df.Affiliation))
```

```python
len(set(df['Last name']))
```

```python
affils = df['Affiliation'].drop_duplicates(keep='first').reset_index(drop=True)
affils
```

```python
idx = df.Affiliation.apply(lambda x: x==affils)
idx
```

## add numbered affiliation

```python
for i in range(idx.shape[1]):
    df.loc[idx[i],'Num']=i+1
```

```python
df.to_csv('toi179_authors.csv', index=False)
```

## format author

```python
lastnames = df['Last name'].drop_duplicates()
firstnames = df['First name'].apply(lambda x: [i[0] for i in x.split()])

text = []
for ln in lastnames:
    idx= df['Last name']==ln
    fn = ".".join(firstnames[idx].values[0])
    nums = sorted(set(df.loc[idx,'Num']))
    nums = ",".join([str(i) for i in nums])
    orcid = df.loc[idx,'ORCID'].iloc[0]
    if len(str(orcid))>3:
        t = f"{fn}~{ln}\orcid"+"{"+f"{orcid}"+"}$^{"+f"{nums}"+"}$,"
    else:
        t = f"{fn}~{ln}$^"+"{"+nums+"}$,"  
    print(t)
    text.append(t)
```

```python
import numpy as np

np.savetxt('authors.txt', text, fmt='%s')
```
